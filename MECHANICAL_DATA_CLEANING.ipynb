```python
import docx

def read_word_file_with_delimiter(filename, delimiter, line_count):
    doc = docx.Document(filename)
    content = []
    current_lines = []

    for paragraph in doc.paragraphs:
        text = paragraph.text.strip()

        if len(current_lines) < line_count:
            current_lines.append(text)
        else:
            content.append(' '.join(current_lines))
            current_lines = [text]

    if current_lines:
        content.append(' '.join(current_lines))

    final_content = []
    for item in content:
        lines = item.split(delimiter)
        final_content.extend(lines)

    return final_content

file_path = 'MechanicalDictionary.xlsx'
delimiter = '@.'
line_count = 3
file_content = read_word_file_with_delimiter(file_path, delimiter, line_count)
file_content
```

```python
import re
# Regular expression pattern to match the capital letter at the start of the response
pattern = re.compile(r'(?<=\s)([A-Z].*?\.)')

# Define two empty lists to store the data for Tag and Response columns
tag_list = []
response_list = []

# Loop through each sentence and separate them into respective columns
for sentence in df['Paragraph']:
    if isinstance(sentence, str):
        match = pattern.search(sentence)
        if match:
            tag_list.append(sentence[:match.start()].strip())
            response_list.append(match.group(0).strip())
        else:
            tag_list.append(sentence.strip())
            response_list.append("")
    else:
        tag_list.append("")
        response_list.append("")



# Create new columns "Tag" and "Response" in the DataFrame
df['Tag'] = tag_list

```

```python
import re

def separate_tag_response(sentence):
    # Define the regular expression pattern for splitting
    pattern = r'(?<=[a-z])(?=[A-Z()])'

    # Find the split position using the regular expression pattern
    match = re.search(pattern, sentence)

    if match:
        split_position = match.start()
        tag = sentence[:split_position].strip()
        response = sentence[split_position:].strip()
    else:
        # If no uppercase letters or brackets found, assume the whole sentence as the response
        tag = ""
        response = sentence.strip()

    return tag, response

i=0
for i in range(0,5889):
  senten=str(df['Paragraph'].loc[df['Paragraph'].index[i]])
  rel,pel=separate_tag_response(senten)

```

```python
import re

def find_position_before_capital(sentence):
    # Regular expression to find the position before a capital letter (except the first one)
    pattern = r'(?<!^)[a-z](?=[A-Z])'

    # Search for the first match in the sentence
    match = re.search(pattern, sentence)

    if match:
        return match.start()
    else:
        return None

i=0
purr=[]
for i in range(0,5890):
  sen=str(df['Paragraph'].loc[df['Paragraph'].index[i]])
  purr.append(find_position_before_capital(sen))
```
